```python

file_name="C:/Users/HP i5/Downloads/framingham.csv"
data = pd.read_csv(file_name)

#drop education since it is unwanted
data.drop(['education'],axis=1,inplace=True)

data
```

```python
counts = data['TenYearCHD'].value_counts()

#data.describe()
data.isna().sum()
total_null_values = data.isnull().sum().sort_values(ascending = False)
percentage = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending = False)
percentage
```

```python
med_g = data['glucose'].median()
med_c = data['BPMeds'].median()
med_d = data['totChol'].median()
med_e = data['cigsPerDay'].median()
med_f = data['BMI'].median()
med_g = data['heartRate'].median()

data['glucose'].fillna(med_g, inplace=True)
data['BPMeds'].fillna(med_c, inplace=True)
data['totChol'].fillna(med_d, inplace=True)
data['cigsPerDay'].fillna(med_e, inplace=True)
data['BMI'].fillna(med_f, inplace=True)
data['heartRate'].fillna(med_g, inplace=True)

data.isna().sum()
total_null_values = data.isnull().sum().sort_values(ascending = False)
percentage = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending = False)
percentage
```

```python
#Adding a constant
from statsmodels.tools import add_constant as add_constant
data_constant = add_constant(data)
data_constant.head()
```

```python
import statsmodels.api as sm
import scipy.stats as st
import matplotlib.pyplot as plt

st.chisqprob = lambda chisq, df: st.chi2.sf(chisq, df)
cols=data_constant.columns[:-1]
model=sm.Logit(data.TenYearCHD,data_constant[cols])
result=model.fit()
result.summary()

```

```python
def back_feature_elem (data_frame,dep_var,col_list):
    while len(col_list)>0 :
        model=sm.Logit(dep_var,data_frame[col_list])
        result=model.fit(disp=0)
        largest_pvalue=round(result.pvalues,3).nlargest(1)
        if largest_pvalue[0]<(0.05):
            return result
            break
        else:
            col_list=col_list.drop(largest_pvalue.index)

result=back_feature_elem(data_constant,data.TenYearCHD,cols)
```

```python
#Interpreting the results: Odds Ratio, Confidence Intervals and Pvalues

params = np.exp(result.params)
conf = np.exp(result.conf_int())
conf['OR'] = params
pvalue=round(result.pvalues,3)
conf['pvalue']=pvalue
conf.columns = ['CI 95%(2.5%)', 'CI 95%(97.5%)', 'Odds Ratio','pvalue']
print ((conf))
```

```python
from sklearn.model_selection import train_test_split
seen_data, unseen_data = train_test_split(data, test_size=0.3, random_state=42)
unseen_data
```

```python
#Splitting data to train and test split


new_features=seen_data[['age','male','cigsPerDay','totChol','sysBP','glucose','TenYearCHD']]
print(new_features)
x=new_features.iloc[:,:-1]
y=new_features.iloc[:,-1]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=5)

```

1. Logistic Regression


```python
from sklearn import metrics
import sklearn
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_train, y_train)
lr_predict = lr.predict(x_test)
lr_conf_matrix = metrics.confusion_matrix(y_test, lr_predict)
lr_acc_score = sklearn.metrics.accuracy_score(y_test, lr_predict)
print("confussion matrix")
print(lr_conf_matrix)
print("\n")
print("Accuracy of Logistic Regression:",lr_acc_score,'\n')
print(metrics.classification_report(y_test,lr_predict))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)

# Creating the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}

# Instantiating logistic regression classifier
logreg = LogisticRegression(max_iter=1000)

# Instantiating the GridSearchCV object
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)

# Fit the model
logreg_cv.fit(x_train_scaled, y_train)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
```

```python
# new data

new_features=unseen_data[['age','male','cigsPerDay','totChol','sysBP','glucose','TenYearCHD']]
print(new_features)
x_unseen=new_features.iloc[:,:-1]
y_unseen=new_features.iloc[:,-1]

logreg=LogisticRegression()
logreg.fit(x_unseen,y_unseen)
y_pred=logreg.predict(x_unseen)
lr_acc_score = sklearn.metrics.accuracy_score(y_unseen, y_pred)
confusion_matrix = metrics.confusion_matrix(y_unseen,y_pred)
print(confusion_matrix)
print("The accuracy of the Logistic Regression is")
print(sklearn.metrics.accuracy_score(y_unseen,y_pred))
print(metrics.classification_report(y_unseen,y_pred))
```

2. KNN


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train, y_train)
knn_predicted = knn.predict(x_test)
knn_conf_matrix = metrics.confusion_matrix(y_test, knn_predicted)
knn_acc_score = accuracy_score(y_test, knn_predicted)
print("confussion matrix")
print(knn_conf_matrix)
print("\n")
print("Accuracy of K-NeighborsClassifier:",knn_acc_score,'\n')
print(classification_report(y_test,knn_predicted))
```
